## 显示图像的方法
如果用plt.show()画图，那么程序会停止执行之后的程序
（不知道是否会写进去，但起码后面加在这个fig上的东西不会显示。可以理解成fig已经关掉了，写不进去了）

为了交互式画图，应该在作图之前先写  
`plt.ion()
fig, ax = ....
ax.text ....
ax.annotate....
plt.ioff()`

注意不要主动关闭plot，如果关了，plt.show()也不会再出现。直接打plt.ioff()然后接plt.show()会显示最终图像，即关交互之前的最后图像。

这是在ipython中的显示方法。在jupyter中，一个cell放这个figure所有的信息，然后直接run，就可以显示最终图像，如果要交互式，还要研究，目前我用的方法是直接改cell来改图像显示。

## Figure 和 Axe 的区别
永远用以下方式画图：  
`fig, ax = plt.subplots(2, 1, figsize = (14, 7))
 ax.plot(x, y)
 ax.title(...)
 ax.xaxis....
 ....`
 
这么做的目的是，区分figure和axe的不同，figure是画布，是生长的图片。而ax才是图，在图上改动东西才是规范的做法。
虽然plt.plot(x, y)也可以作图，但是不正规，不应该这么写。

## Annotate各种参数
annotate是在axe上加标注 一般写成  
`ax.annotate("test", xy=(1, 1), xycoords="data", xytext=(0.8, 0.9), textcoords = "axesfraction", 
arrowprops=dict(facecolor='black', shrink=0.05), bbox=dict(boxstyle = "round", fc = "w"), va = "center", ha = "center")`

xy是指向点的坐标，它用的坐标系xycoords是data，即默认正交坐标系。xytext是标注所在的位置，它用的坐标系是axesfraction，即以在axe上的相对位置来定，(0,0)是axe左下角，(1,1)是axe右上角，这个例子中是在偏右上角。  
arrowprops是指用箭头指向点。bbox是指标注用box框起来，这两个都有不同的参数设置，但都是用dict这个形式在写。  
va是指竖直对齐方向，center就是说生成的这个box的中心在xytext位置，而ha是指水平对齐方向。





```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

A = np.arange(1,5)
B = A**2
C = A**3

fig, ax = plt.subplots(figsize=(14,7))
ax.plot(A,B)
ax.plot(B,A)

ax.set_title('Title',fontsize=18)
ax.set_xlabel('xlabel', fontsize=18,fontfamily = 'sans-serif',fontstyle='italic')
ax.set_ylabel('ylabel', fontsize='x-large',fontstyle='oblique')
ax.legend()
ax.set_aspect('equal') 
ax.minorticks_on() 
ax.set_xlim(0,16) 
ax.grid(which='minor', axis='both')
ax.xaxis.set_tick_params(rotation=45,labelsize=18,colors='w') 
start, end = ax.get_xlim() 
ax.xaxis.set_ticks(np.arange(start, end,1)) 
ax.yaxis.tick_right()

```

```python
fig, ax = plt.subplots()
an1 = ax.annotate("Test 1", xy=(0.5, 0.5), xycoords="data",
                  va="center", ha="left",
                  bbox=dict(boxstyle="round", fc="w"))
```
